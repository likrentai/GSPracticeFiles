## STAGE ONE: Data Cleaning
Original Dataset: DATA-Japan_Universities.csv

EXCEL Spreadsheet Steps:
1. Find & select the column 'code'.
2. Insert 1 column to the right → Name column header: AREA
3. In this new column on the first cell, enter: =MID(B2,3,2)
4. Press: Ctrl+Shift+↓ (to select all the cells in the column w/o header)
5. Press: Ctrl+d (to apply the formula to all cells)
6. Insert 1 column to the right → Name column header: FIPS
7. In this new column on the first cell, enter: =TEXT(C2,"JP-00")
8. Press: Ctrl+Shift+↓ (to select all the cells in the column w/o header)
9. Press: Ctrl+d (to apply the formula to all cells)
10. Remove unwanted columns
11. File export and saved as **DATA-JP-UNI-CLEAN1.xlsx**
